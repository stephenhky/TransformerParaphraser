```python
from transformers import (
    AutoModelForCausalLM,
    AutoConfig,
    Trainer,
    AutoTokenizer,
    TextDataset,
    DataCollatorForLanguageModeling,
    TrainingArguments
)
```

```python
def modelTrainer(text_path, epochs, output_dir, model='gpt2', tokenizer='gpt2', batch_size=8):
    model = AutoModelForCausalLM.from_pretrained(model)
    tokenizer = AutoTokenizer.from_pretrained(tokenizer)
    data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
    train_dataset = TextDataset(
        tokenizer=tokenizer,
        file_path=text_path,
        block_size=256
    )
    training_args = TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=epochs,
        per_gpu_train_batch_size=batch_size,
        warmup_steps=500,
        save_steps=2000,
        logging_steps=10
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=train_dataset
    )

    trainer.train()
    trainer.save_model()
```

```python
acl_train_path = '/content/drive/MyDrive/Paraphrasing/ParaSci/data/ACL_train.tsv'
acl_output_dir = '/content/drive/MyDrive/Paraphrasing/ParaSci/model/ACL'
```

```python
modelTrainer(acl_train_path, 100, acl_output_dir)
```

```python
arxiv_train_path = '/content/drive/MyDrive/Paraphrasing/ParaSci/data/arXiv_train.tsv'
arxiv_output_dir = '/content/drive/MyDrive/Paraphrasing/ParaSci/model/arXiv'
```

```python
modelTrainer(arxiv_train_path, 10, arxiv_output_dir)
```
