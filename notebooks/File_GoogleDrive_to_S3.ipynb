```python
import os
from glob import glob

import boto3
from google.colab import userdata
```

```python
s3_client = boto3.client(
    's3',
    aws_access_key_id=userdata.get('AWS_STEPHENHO_SERVERLESS_ACC_ACCESS_TOKEN'),
    aws_secret_access_key=userdata.get('AWS_STEPHENHO_SERVERLESS_ACC_SECRET_ACCESS_KEY')
)
```

```python
def upload_directory_to_s3(directory, s3_bucket, s3dir):
  print(">> Directory: "+directory)
  for path in glob(os.path.join(directory, '*')):
    print(">>> Path: "+path)
    if os.path.isdir(path):
      basedir = os.path.basename(path)
      upload_directory_to_s3(path, s3_bucket, os.path.join(s3dir, basedir))
    else:
      basename = os.path.basename(path)
      print(basename+'-->'+os.path.join(s3dir, basename))
      s3_client.upload_file(
          path,
          'paraphrase-parasci',
          os.path.join(s3dir, basename)
      )
```

```python
upload_directory_to_s3(
    '/content/drive/MyDrive/Paraphrasing/ParaSci/model/ACL',
    'paraphrase-parasci',
    'models/ACL'
)
```
